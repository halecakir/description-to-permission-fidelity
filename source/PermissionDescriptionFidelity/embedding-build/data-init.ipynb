```python
import csv
import demoji
demoji.download_codes()

from utils.nlp_utils import NLPUtils
%aimport utils.nlp_utils
```

```python
def remove_emoji(line):
    return demoji.replace(line, repl="")

def is_ascii(s):
    return all(ord(c) < 128 for c in s)

def has_digit(string):
    return any(char.isdigit() for char in string)
```

```python
def preprocess_description(description):
    text_data = []
    text = NLPUtils.remove_hyperlinks(description)
    for line in NLPUtils.sentence_tokenization(text):
        line = remove_emoji(line)
        line = NLPUtils.punctuation_removal(line)
        tokenized_words = NLPUtils.word_tokenization_nltk(line)
        words = NLPUtils.stopword_elimination(tokenized_words)

        sentence = []
        for word in words:
            if is_ascii(word) and (not has_digit(word)):
                sentence.append(word.lower())
        if len(sentence) > 1:
            text_data.append(sentence)
    return text_data
```

```python
apps_path = "/home/huseyinalecakir/Security/data/big_processed/apps.csv"

with open(apps_path) as stream:
    reader = csv.reader(stream)
    header = next(reader)
    data = []
    
    for i, row in enumerate(reader):
        if i % 1000 == 0:
            print("App id {}".format(i))
        text = row[1]
        data.extend(preprocess_description(text))
```
